```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows',None)
pd.set_option('display.float_format', '{:.1f}'.format)
```

## Reading Survey_Result_CSV file.

```python
main_data_2023=pd.read_csv("survey_results_public.csv")
main_data_2023
```

# Copying main-data into another variable.

```python
main_df_2023=main_data_2023.copy()
main_df_2023
```

## Filtering out the column I might use for EDA. 

```python
df_2023_final=main_df_2023.loc[:,["Age","Employment","RemoteWork","EdLevel","LearnCode","LearnCodeCoursesCert",
                                  "DevType","OrgSize","Country","ConvertedCompYearly",
                                  "LanguageHaveWorkedWith","DatabaseHaveWorkedWith","WebframeHaveWorkedWith",
                                             "PlatformHaveWorkedWith","OpSysPersonal use","OpSysProfessional use",
                                             "SOVisitFreq","WorkExp"]]
df_2023_final
```

# Data Wrangling.

### 1. Renaming  columns

Age -> Age_group    

Edlevel -> Education

RemoteWork -> Work_Place

LearnCode -> Learning_method

LearnCodeCoursesCert -> Online_learning_platform

ConvertedCompYearly -> Salary_annually(in US Dollars)

OpSysPersonal use -> OS_for_personal_use 

OpSysProfessional use -> OS_for_professional_use

SOVisitFreq -> How_often_you_visit_SO

```python
df_2023_final.rename(columns={"Age":"Age_group","EdLevel":"Education","RemoteWork":"Workplace",
                             "LearnCode":"Learning_method",
                              "LearnCodeCoursesCert": "Online_learning_platform",
                             "ConvertedCompYearly":"Salary_annually(in US Dollar)",
                              "OpSysPersonal use":"OS_for_personal_use","OpSysProfessional use":"OS_for_professional_use",
                              "SOVisitFreq":"How_often_you_visit_SO"
                             },inplace=True)
```

```python
df_2023_final.head()
```

# 2.Cleaning column 

###  Age column.

18-24 years old -> 18-24 

Prefer not to sa -> N/A

```python
df_2023_final['Age_group']=df_2023_final["Age_group"].str.split('y').str[0]

df_2023_final["Age_group"]=df_2023_final["Age_group"].replace('Prefer not to sa','N/A')
```

```python
df_2023_final.tail()
```

.

## Education column.

Bachelor’s degree (B.A., B.S., B.Eng., etc.)   -> Bachelor’s degree

```python
df_2023_final["Education"]=df_2023_final["Education"].str.split("(").str[0]
```

```python
df_2023_final.head()
```

## Country column.

### Shortening name of countries.

"United States of America" : "USA"

"United Kingdom of Great Britain and Northern I..." : "UK_&_NI"

"Iran, Islamic Republic of..." : "Iran"

"United Arab Emirates" : "UAE"

"Republic of Korea" : "South_Korea"

"South Korea" : "South_Korea"

"The former Yugoslav Republic of Macedonia"  : "North_Macedonia"

"Venezuela, Bolivarian Republic of..." : "Venezuela"

"Congo, Republic of the..." : "Congo"

"Lao People's Democratic Republic" : "Laos"

"North Korea" : "North_Korea"

"Democratic People's Republic of Korea" : "North_Korea"

"Democratic Republic of the Congo"  : "DRC"

"Russian Federation" : "Russia" 

"Libyan Arab Jamahiriya" : "Libya"

"United Republic of Tanzania":"Tanzania"

```python
countries_name_to_shorten={"United States of America" : "USA",
                          "United Kingdom of Great Britain and Northern I..." : "UK_&_NI",
                          "United Kingdom of Great Britain and Northern Ireland" : "UK_&_NI",
                            "Iran, Islamic Republic of..." : "Iran","United Arab Emirates" : "UAE",
                          "Republic of Korea" : "South_Korea","South Korea" : "South_Korea",
                          "The former Yugoslav Republic of Macedonia"  : "North_Macedonia",
                          "Venezuela, Bolivarian Republic of..." : "Venezuela","Congo, Republic of the..." : "Congo",
                          "Lao People's Democratic Republic" : "Laos","North Korea" : "North_Korea",
                          "Democratic People's Republic of Korea" : "North_Korea",
                          "Democratic Republic of the Congo"  : "DRC","Russian Federation" : "Russia" ,
                          "Libyan Arab Jamahiriya" : "Libya","United Republic of Tanzania":"Tanzania"}
                          
```

```python
for i in countries_name_to_shorten.items():
    df_2023_final["Country"]=df_2023_final["Country"].replace(i[0],i[1])
```

```python
df_2023_final.head(100)
```

.

## 3.Converting all multiple values separated by semicolon(';') to list.

```python
column_name=["Learning_method","LanguageHaveWorkedWith","DatabaseHaveWorkedWith",
             "Online_learning_platform",
            "WebframeHaveWorkedWith","PlatformHaveWorkedWith","OS_for_personal_use","OS_for_professional_use"]
```

```python
for col_name in column_name:
    df_2023_final[col_name]=df_2023_final[col_name].str.split(";")
```

## Salary Analysis

Removing outlier in salary column for better analysis.

```python
q1=df_2023_final["Salary_annually(in US Dollar)"].quantile(0.25)
q3=df_2023_final["Salary_annually(in US Dollar)"].quantile(0.75)

IQR = q3-q1

Lf=q1-(1.5*IQR)
uf=q3+(1.5*IQR)

df_after_filtering=df_2023_final[(df_2023_final["Salary_annually(in US Dollar)"] > Lf) & 
                                    (df_2023_final["Salary_annually(in US Dollar)"] < uf)]
```

```python
plt.figure(figsize=(10,3))
sns.scatterplot(x=df_after_filtering["Salary_annually(in US Dollar)"],y=df_after_filtering["Education"])
plt.title("Relation b/w education & Salary")
plt.show()
```

People with graduation degree or master degree or professional degree are sometimes earning same amount.

So degree will  help you with foot in the door, rest its all upto skills.

## What OS have repondents worked with at personal and professional place?

```python
os_personal=pd.DataFrame(df_2023_final["OS_for_personal_use"].apply(pd.Series).stack().value_counts())
os_personal.reset_index(inplace=True)
os_personal.rename(columns={"index":"OS",0:"Personal"},inplace=True)
```

```python
os_professional=pd.DataFrame(df_2023_final["OS_for_professional_use"].apply(pd.Series).stack().value_counts())
os_professional.reset_index(inplace=True)
os_professional.rename(columns={"index":"OS",0:"Professional"},inplace=True)
```

```python
os_df=pd.merge(os_personal,os_professional,how="inner",on="OS")
```

```python
plotting_os=pd.melt(os_df,id_vars="OS",value_vars=["Personal","Professional"])
```

```python
plt.figure(figsize=(10,7))
sns.barplot(y=plotting_os["OS"],x=plotting_os["value"],hue=plotting_os["variable"])
plt.xlabel("Usage Counts")
plt.legend(loc='center right')
plt.title("OS used by respondents for personal & professional use")
plt.show()
```

# Organisation-size wise what is median & highest salary ?

```python
max_sal=pd.DataFrame(df_2023_final.groupby(["OrgSize"])["Salary_annually(in US Dollar)"].max())
max_sal.reset_index(inplace=True)
max_sal.rename(columns={"Salary_annually(in US Dollar)":"Highest_Salary(in_US_Dollar)"},inplace=True)
max_sal.sort_values("Highest_Salary(in_US_Dollar)",ascending=False,inplace=True)
```

```python
median_sal=pd.DataFrame(df_2023_final.groupby(["OrgSize"])["Salary_annually(in US Dollar)"].median())
median_sal.reset_index(inplace=True)
median_sal.rename(columns={"Salary_annually(in US Dollar)":"Median_Salary(in_US_Dollar)"},inplace=True)
median_sal.sort_values("Median_Salary(in_US_Dollar)",ascending=False,inplace=True)
```

```python
sal_bar=pd.merge(median_sal,max_sal,how='inner',on="OrgSize")
```
